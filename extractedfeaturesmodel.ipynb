```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import ReduceLROnPlateau
import pywt  # PyWavelets for wavelet transform

# Device
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Paths
DATA_DIR = "/kaggle/input/movement2/movement"
LABEL_PATH = "/kaggle/input/patient-labels2/patient_labels.csv"

# Load functions
def load_bin_file(filepath):
    data = np.fromfile(filepath, dtype=np.float32)
    assert data.shape[0] == 128832
    return data.reshape(132, 976)

def load_all_files(data_dir):
    X = []
    filepaths = sorted([
        os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith(".bin")
    ])
    print(f"Found {len(filepaths)} files.")
    
    file_indices = list(range(len(filepaths)))
    return filepaths, file_indices

def load_data_by_indices(filepaths, indices):
    """Load only specific files based on indices"""
    X = []
    for idx in tqdm(indices):
        path = filepaths[idx]
        X.append(load_bin_file(path))
    return np.stack(X)

def load_labels(label_path):
    df = pd.read_csv(label_path)
    return df['label'].values.astype(np.float32)

# Normalize function to be applied separately to train and test
def normalize_data(X):
    """Normalize data without leaking statistics between train and test"""
    return (X - X.mean(axis=(0, 2), keepdims=True)) / (X.std(axis=(0, 2), keepdims=True) + 1e-6)

# Feature extraction functions
def extract_statistical_features(X):
    """Extract statistical features from time series data"""
    # X shape: (samples, channels, time_steps)
    features = []
    
    for i in range(X.shape[0]):  # For each sample
        sample_features = []
        
        for j in range(X.shape[1]):  # For each channel
            channel_data = X[i, j, :]
            
            # Basic statistics
            mean = np.mean(channel_data)
            std = np.std(channel_data)
            min_val = np.min(channel_data)
            max_val = np.max(channel_data)
            median = np.median(channel_data)
            
            # Signal characteristics
            rms = np.sqrt(np.mean(channel_data**2))
            peak_to_peak = max_val - min_val
            
            # Shape characteristics
            skewness = np.mean(((channel_data - mean) / (std + 1e-10))**3) if std > 0 else 0
            kurtosis = np.mean(((channel_data - mean) / (std + 1e-10))**4) if std > 0 else 0
            
            # Frequency domain (simple)
            fft_vals = np.abs(np.fft.rfft(channel_data))
            fft_freq = np.fft.rfftfreq(len(channel_data))
            dominant_freq_idx = np.argmax(fft_vals)
            dominant_freq = fft_freq[dominant_freq_idx]
            dominant_power = fft_vals[dominant_freq_idx]
            
            # Energy in different frequency bands
            low_freq_power = np.sum(fft_vals[fft_freq < 0.1]**2)
            mid_freq_power = np.sum(fft_vals[(fft_freq >= 0.1) & (fft_freq < 0.3)]**2)
            high_freq_power = np.sum(fft_vals[fft_freq >= 0.3]**2)
            
            # Combine features
            channel_features = [mean, std, min_val, max_val, median, 
                               rms, peak_to_peak, skewness, kurtosis,
                               dominant_freq, dominant_power,
                               low_freq_power, mid_freq_power, high_freq_power]
            
            sample_features.extend(channel_features)
        
        features.append(sample_features)
    
    return np.array(features)

def extract_wavelet_features(X, wavelet='db4', level=5):
    """Extract wavelet transform features from time series data"""
    features = []
    
    for i in range(X.shape[0]):  # For each sample
        sample_features = []
        
        for j in range(X.shape[1]):  # For each channel
            channel_data = X[i, j, :]
            
            # Perform wavelet decomposition
            coeffs = pywt.wavedec(channel_data, wavelet, level=level)
            
            # Extract features from coefficients
            for k, coeff in enumerate(coeffs):
                # For each decomposition level, extract statistics
                mean = np.mean(coeff)
                std = np.std(coeff)
                energy = np.sum(coeff**2)
                
                sample_features.extend([mean, std, energy])
        
        features.append(sample_features)
    
    return np.array(features)

def extract_cross_channel_features(X):
    """Extract features that capture relationships between channels"""
    features = []
    
    for i in range(X.shape[0]):  # For each sample
        sample = X[i]  # (channels, time_steps)
        
        # Correlation matrix between channels
        corr_matrix = np.corrcoef(sample)
        
        # Extract upper triangular part of correlation matrix (without diagonal)
        upper_tri = corr_matrix[np.triu_indices(corr_matrix.shape[0], k=1)]
        
        # Compute some statistics on correlations
        mean_corr = np.mean(upper_tri)
        std_corr = np.std(upper_tri)
        max_corr = np.max(upper_tri)
        min_corr = np.min(upper_tri)
        
        # Frequency domain coherence (simplified)
        sample_features = [mean_corr, std_corr, max_corr, min_corr]
        
        # Add some key channel pairs (assuming domain knowledge about which channels matter)
        # This would be better if we knew which channels correspond to which sensors
        for c1 in range(0, sample.shape[0], 10):
            for c2 in range(c1+5, min(c1+10, sample.shape[0])):
                if c2 < sample.shape[0]:
                    # Correlation
                    corr = np.corrcoef(sample[c1], sample[c2])[0, 1]
                    
                    # Phase sync (simplified)
                    fft1 = np.fft.rfft(sample[c1])
                    fft2 = np.fft.rfft(sample[c2])
                    phase1 = np.angle(fft1)
                    phase2 = np.angle(fft2)
                    phase_diff = np.mean(np.abs(phase1 - phase2))
                    
                    sample_features.extend([corr, phase_diff])
        
        features.append(sample_features)
    
    return np.array(features)

def preprocess_data(X):
    """Combine multiple feature extraction methods"""
    print("Extracting statistical features...")
    statistical_features = extract_statistical_features(X)
    
    print("Extracting wavelet features...")
    wavelet_features = extract_wavelet_features(X)
    
    print("Extracting cross-channel features...")
    cross_channel_features = extract_cross_channel_features(X)
    
    # Combine all features
    combined_features = np.hstack([
        statistical_features, 
        wavelet_features, 
        cross_channel_features
    ])
    
    print(f"Final feature shape: {combined_features.shape}")
    return combined_features

# Feature importance and selection
def feature_selection(X, y, threshold=0.1):
    """Select important features using Random Forest feature importance"""
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.feature_selection import SelectFromModel
    
    # Train a Random Forest to get feature importances
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # Select features based on importance
    selector = SelectFromModel(rf, threshold=threshold, prefit=True)
    X_selected = selector.transform(X)
    
    # Get feature indices
    selected_indices = selector.get_support(indices=True)
    
    print(f"Selected {X_selected.shape[1]} features out of {X.shape[1]}")
    return X_selected, selected_indices, selector

class ParkinsonClassifier(nn.Module):
    def __init__(self, input_size, hidden_sizes=[256, 128, 64], dropout_rate=0.3, batch_norm=True):
        super(ParkinsonClassifier, self).__init__()
        
        self.batch_norm = batch_norm
        
        # Input layer
        layers = [nn.Linear(input_size, hidden_sizes[0])]
        
        # Hidden layers
        for i in range(len(hidden_sizes)-1):
            if batch_norm:
                layers.extend([
                    nn.BatchNorm1d(hidden_sizes[i]),
                    nn.ReLU(),
                    nn.Dropout(dropout_rate),
                    nn.Linear(hidden_sizes[i], hidden_sizes[i+1])
                ])
            else:
                layers.extend([
                    nn.ReLU(),
                    nn.Dropout(dropout_rate),
                    nn.Linear(hidden_sizes[i], hidden_sizes[i+1])
                ])
        
        # Combine layers into feature extractor
        self.features = nn.Sequential(*layers)
        
        # Output layer
        if batch_norm:
            self.classifier = nn.Sequential(
                nn.BatchNorm1d(hidden_sizes[-1]),
                nn.ReLU(),
                nn.Dropout(dropout_rate),
                nn.Linear(hidden_sizes[-1], 1)
            )
        else:
            self.classifier = nn.Sequential(
                nn.ReLU(),
                nn.Dropout(dropout_rate),
                nn.Linear(hidden_sizes[-1], 1)
            )
        
        # Initialize weights
        self._initialize_weights()
    
    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
    
    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)

# Early stopping
class EarlyStopping:
    def __init__(self, patience=10, min_delta=0, restore_best_weights=True):
        self.patience = patience
        self.min_delta = min_delta
        self.restore_best_weights = restore_best_weights
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.best_weights = None
    
    def __call__(self, score, model):
        if self.best_score is None:
            self.best_score = score
            self.best_weights = model.state_dict().copy()
        elif score <= self.best_score + self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.best_weights = model.state_dict().copy()
            self.counter = 0
        
        return self.early_stop

# Training function
def train_model(X_train, y_train, X_val, y_val, hidden_sizes=[256, 128, 64], 
               dropout_rate=0.3, lr=0.001, batch_size=32, weight_decay=1e-5, 
               patience=15, num_epochs=200):
    
    # Check if batch size is too small for batch norm
    use_batch_norm = batch_size > 1
    
    # Convert to PyTorch tensors
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
    X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
    y_val_tensor = torch.tensor(y_val, dtype=torch.float32).unsqueeze(1)
    
    # Create data loaders - ensure at least 2 samples per batch for training if using batch norm
    if use_batch_norm:
        train_batch_size = max(2, batch_size)
    else:
        train_batch_size = batch_size
        
    train_loader = DataLoader(TensorDataset(X_train_tensor, y_train_tensor), 
                             batch_size=train_batch_size, shuffle=True)
    val_loader = DataLoader(TensorDataset(X_val_tensor, y_val_tensor), 
                           batch_size=batch_size)
    
    # Class weights for imbalanced data
    num_positives = (y_train_tensor == 1).sum().item()
    num_negatives = (y_train_tensor == 0).sum().item()
    pos_weight = torch.tensor([num_negatives / num_positives]).to(DEVICE)
    
    # Initialize model with appropriate batch norm setting
    model = ParkinsonClassifier(
        input_size=X_train.shape[1],
        hidden_sizes=hidden_sizes,
        dropout_rate=dropout_rate,
        batch_norm=use_batch_norm
    ).to(DEVICE)
    
    # Loss and optimizer
    criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=5, verbose=True)
    early_stopping = EarlyStopping(patience=patience)
    
    # Training history
    history = {'train_loss': [], 'val_loss': [], 'val_accuracy': [], 'val_auc': []}
    
    # Training loop
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0
        
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            
            # Skip single-sample batches if using batch norm
            if use_batch_norm and X_batch.size(0) < 2:
                continue
                
            # Forward pass
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            
            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            
            # Gradient clipping
            nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            # Update weights
            optimizer.step()
            
            train_loss += loss.item()
        
        avg_train_loss = train_loss / max(1, len(train_loader))
        history['train_loss'].append(avg_train_loss)
        
        # Validation phase
        model.eval()
        val_loss = 0
        all_preds, all_probs, all_labels = [], [], []
        
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
                
                # Skip single-sample batches if using batch norm
                if use_batch_norm and X_batch.size(0) < 2:
                    continue
                
                # Forward pass
                outputs = model(X_batch)
                loss = criterion(outputs, y_batch)
                
                # Save predictions
                probs = torch.sigmoid(outputs).cpu().numpy()
                preds = (probs > 0.5).astype(int)
                
                all_probs.extend(probs)
                all_preds.extend(preds)
                all_labels.extend(y_batch.cpu().numpy())
                
                val_loss += loss.item()
        
        # Calculate metrics
        avg_val_loss = val_loss / max(1, len(val_loader))
        
        if len(all_labels) > 0:
            accuracy = accuracy_score(all_labels, all_preds)
            try:
                auc = roc_auc_score(all_labels, all_probs)
            except:
                auc = 0.5
        else:
            accuracy = 0
            auc = 0.5
        
        # Update history
        history['val_loss'].append(avg_val_loss)
        history['val_accuracy'].append(accuracy)
        history['val_auc'].append(auc)
        
        # Print progress
        print(f"Epoch {epoch+1}/{num_epochs} - "
              f"Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}, "
              f"Accuracy: {accuracy:.4f}, AUC: {auc:.4f}")
        
        # Learning rate scheduling
        scheduler.step(accuracy)
        
        # Early stopping
        if early_stopping(accuracy, model):
            print(f"Early stopping at epoch {epoch+1}")
            model.load_state_dict(early_stopping.best_weights)
            break
    
    # Final evaluation
    model.eval()
    all_preds, all_probs, all_labels = [], [], []
    
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(DEVICE)
            
            # Skip single-sample batches if using batch norm
            if use_batch_norm and X_batch.size(0) < 2:
                continue
                
            outputs = model(X_batch)
            probs = torch.sigmoid(outputs).cpu().numpy()
            preds = (probs > 0.5).astype(int)
            all_probs.extend(probs)
            all_preds.extend(preds)
            all_labels.extend(y_batch.cpu().numpy())
    
    # Calculate final metrics
    if len(all_labels) > 0:
        accuracy = accuracy_score(all_labels, all_preds)
        precision = precision_score(all_labels, all_preds)
        recall = recall_score(all_labels, all_preds)
        f1 = f1_score(all_labels, all_preds)
        try:
            auc = roc_auc_score(all_labels, all_probs)
        except:
            auc = 0.5
    else:
        accuracy = precision = recall = f1 = 0
        auc = 0.5
    
    print("\nFinal Model Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"AUC: {auc:.4f}")
    
    return model, history, {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'auc': auc
    }

# Add this new helper function to check class distribution
def check_class_distribution(y_train, y_val):
    """Check if both sets have at least one sample from each class"""
    train_classes = set(y_train)
    val_classes = set(y_val)
    
    # Check if all classes are present in both sets
    all_classes = set([0, 1])  # Binary classification
    
    if len(train_classes) < len(all_classes) or len(val_classes) < len(all_classes):
        print("Warning: Not all classes present in train or validation set")
        return False
    
    return True

# Evaluate model on held-out test set
def evaluate_model(model, X_test, y_test, batch_size=32):
    """Evaluate model on held-out test set for final reporting"""
    model.eval()
    
    # Convert to PyTorch tensors
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)
    
    # Create data loader
    test_loader = DataLoader(TensorDataset(X_test_tensor, y_test_tensor), 
                            batch_size=batch_size)
    
    # Predictions
    all_preds, all_probs, all_labels = [], [], []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(DEVICE)
            outputs = model(X_batch)
            probs = torch.sigmoid(outputs).cpu().numpy()
            preds = (probs > 0.5).astype(int)
            all_probs.extend(probs)
            all_preds.extend(preds)
            all_labels.extend(y_batch.cpu().numpy())
    
    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds)
    recall = recall_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds)
    try:
        auc = roc_auc_score(all_labels, all_probs)
    except:
        auc = 0.5
    
    print("\nFinal Test Set Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"AUC: {auc:.4f}")
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'auc': auc
    }

# Function to train a final model using cross-validation for hyperparameter tuning
def train_final_model(X_train, y_train, params, k_folds=5):
    """Train final model using k-fold cross-validation to ensure robustness"""
    print("\nTraining final model using {}-fold cross-validation...".format(k_folds))
    
    # Setup k-fold cross validation
    kfold = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    
    # Store metrics for each fold
    fold_metrics = []
    
    # Cross-validation process
    for fold, (train_idx, val_idx) in enumerate(kfold.split(X_train)):
        print(f"\nFold {fold+1}/{k_folds}")
        
        # Split data
        X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
        y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
        
        # Train model for this fold
        model, _, metrics = train_model(
            X_train_fold, y_train_fold, X_val_fold, y_val_fold,
            hidden_sizes=params['hidden_sizes'],
            dropout_rate=params['dropout_rate'],
            lr=params['lr'],
            batch_size=params['batch_size'],
            weight_decay=params['weight_decay'],
            patience=params['patience']
        )
        
        fold_metrics.append(metrics)
    
    # Print averaged results from cross-validation
    print("\nCross-validation results:")
    avg_metrics = {metric: np.mean([fold[metric] for fold in fold_metrics]) for metric in fold_metrics[0]}
    for metric, value in avg_metrics.items():
        print(f"Average {metric}: {value:.4f}")
    
    # Now train the final model on the entire training set
    print("\nTraining final model on full training dataset...")
    
    # Create a validation split from the training data for early stopping (10% of training data)
    X_train_final, X_val_final, y_train_final, y_val_final = train_test_split(
        X_train, y_train, test_size=0.1, random_state=42, stratify=y_train
    )
    
    final_model, _, _ = train_model(
        X_train_final, y_train_final, X_val_final, y_val_final,
        hidden_sizes=params['hidden_sizes'],
        dropout_rate=params['dropout_rate'],
        lr=params['lr'],
        batch_size=params['batch_size'],
        weight_decay=params['weight_decay'],
        patience=params['patience']
    )
    
    return final_model, avg_metrics

# Plot training history
def plot_history(history):
    plt.figure(figsize=(12, 4))
    
    # Plot training & validation loss
    plt.subplot(1, 3, 1)
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot validation accuracy
    plt.subplot(1, 3, 2)
    plt.plot(history['val_accuracy'], label='Accuracy')
    plt.title('Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    
    # Plot validation AUC
    plt.subplot(1, 3, 3)
    plt.plot(history['val_auc'], label='AUC')
    plt.title('Validation AUC')
    plt.xlabel('Epochs')
    plt.ylabel('AUC')
    
    plt.tight_layout()
    plt.show()

# Main function
def main():
    print("Loading data file paths...")
    filepaths, file_indices = load_all_files(DATA_DIR)
    
    # Load labels
    y = load_labels(LABEL_PATH)
    
    # Print class distribution
    print("Class distribution:")
    unique, counts = np.unique(y, return_counts=True)
    for value, count in zip(unique, counts):
        print(f"Class {int(value)}: {count} ({count/len(y)*100:.2f}%)")
    
    # First, split the indices to ensure proper train-test separation
    # This way we'll only load the data we need for each set
    train_indices, test_indices, y_train, y_test = train_test_split(
        file_indices, y, test_size=0.2, random_state=42, stratify=y
    )
    
    print(f"Training set size: {len(train_indices)}, Test set size: {len(test_indices)}")
    
    # Load and process training data
    print("\nLoading and processing training data...")
    X_train_raw = load_data_by_indices(filepaths, train_indices)
    
    # Normalize training data
    X_train_normalized = normalize_data(X_train_raw)
    
    # Extract features from training data
    X_train_features = preprocess_data(X_train_normalized)
    
    # Feature selection on training data only
    X_train_selected, selected_indices, selector = feature_selection(X_train_features, y_train, threshold='median')
    
    # Make sure we have at least 8 samples (minimum to allow 2 samples per batch in 4 folds)
    min_required_samples = 8
    if len(X_train_selected) < min_required_samples:
        print(f"Warning: Not enough training samples. Need at least {min_required_samples}, got {len(X_train_selected)}")
        print("Using a simpler train/test approach instead of cross-validation.")
        
        # Split training data for initial model - use a larger validation set
        X_train, X_val, y_train_split, y_val = train_test_split(
            X_train_selected, y_train, test_size=0.3, random_state=42, stratify=y_train
        )
        
        # Ensure we have at least 2 samples per class
        if check_class_distribution(y_train_split, y_val):
            # Train a simple model
            print("\nTraining a simple model...")
            model, history, metrics = train_model(
                X_train, y_train_split, X_val, y_val,
                hidden_sizes=[64, 32],  # Smaller model
                dropout_rate=0.3,
                lr=0.001,
                batch_size=2,  # Minimum batch size
                weight_decay=1e-4,
                patience=10
            )
            
            # Define simplified parameters
            best_params = {
                'hidden_sizes': [64, 32],
                'dropout_rate': 0.3,
                'lr': 0.001,
                'batch_size': 2,
                'weight_decay': 1e-4,
                'patience': 10
            }
            
            # Skip cross-validation and use this model as final
            final_model = model
            cv_metrics = metrics
        else:
            # If we can't even get a good train/val split, use a dummy model
            print("Unable to create valid train/val split. Using a dummy classifier.")
            from sklearn.dummy import DummyClassifier
            dummy = DummyClassifier(strategy='most_frequent').fit(X_train_selected, y_train)
            
            # Create a wrapper class
            class DummyWrapper(nn.Module):
                def __init__(self, dummy_model):
                    super().__init__()
                    self.dummy_model = dummy_model
                
                def forward(self, x):
                    return torch.tensor([[0.0]]).to(DEVICE)  # Placeholder
            
            final_model = DummyWrapper(dummy)
            cv_metrics = {'accuracy': dummy.score(X_train_selected, y_train), 
                         'precision': 0, 'recall': 0, 'f1': 0, 'auc': 0.5}
    else:
        # We have enough samples - proceed with standard approach
        
        # Split training data for initial model
        X_train, X_val, y_train_split, y_val = train_test_split(
            X_train_selected, y_train, test_size=0.25, random_state=42, stratify=y_train
        )
        
        # Train the initial model for hyperparameter tuning
        print("\nTraining and evaluating initial model...")
        model, history, metrics = train_model(
            X_train, y_train_split, X_val, y_val,
            hidden_sizes=[512, 256, 128],
            dropout_rate=0.4,
            lr=0.0005,
            batch_size=16,
            weight_decay=1e-4,
            patience=15
        )
        
        # Plot training history
        plot_history(history)
        
        # Define best parameters based on initial training
        best_params = {
            'hidden_sizes': [512, 256, 128],
            'dropout_rate': 0.4,
            'lr': 0.0005,
            'batch_size': max(2, min(16, len(X_train_selected) // 4)),  # Adjust batch size based on data size
            'weight_decay': 1e-4,
            'patience': 15
        }
        
        # Train final model using cross-validation on the training set
        final_model, cv_metrics = train_final_model(X_train_selected, y_train, best_params, k_folds=5)
    
    # Now, load and process test data
    print("\nLoading and processing test data...")
    X_test_raw = load_data_by_indices(filepaths, test_indices)
    
    # Normalize test data separately
    X_test_normalized = normalize_data(X_test_raw)
    
    # Extract features from test data
    X_test_features = preprocess_data(X_test_normalized)
    
    # Apply the same feature selection to the test data
    X_test_selected = selector.transform(X_test_features)
    
    # Evaluate the final model on the completely held-out test set
    test_metrics = evaluate_model(final_model, X_test_selected, y_test)
    
    # Save the final model and related components
    torch.save({
        'model_state_dict': final_model.state_dict(),
        'feature_selector': selector,
        'model_params': best_params,
        'cv_metrics': cv_metrics,
        'test_metrics': test_metrics,
        'input_shape': X_train_selected.shape[1]
    }, 'parkinson_classifier_final.pth')
    
    print("\nFinal model saved as 'parkinson_classifier_final.pth'")
    print("\nDone!")

if __name__ == "__main__":
    main()
```
